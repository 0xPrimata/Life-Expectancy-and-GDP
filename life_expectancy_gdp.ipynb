# **Life Expectancy and GDP**

  ### Here we observe the correlation between the GDP and the Life Expectancy comparing countries from different regions and distinct backgrounds.
  ### Necessary tools:


```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('all_data.csv')
print(df.head())
print(df.Country.value_counts())
```

### From the data inspection above we can affirm that:
### The dataset contains data from 6 countries: Chile, Zimbabwe, China, Germany, Mexico and United States of America.
### It has 4 columns: Country, Year, Life expectancy at birth (years) and GDP.
### The data ranges from the year 2000 to 2015.
### First we could check an overall trend among countries:


```python
sns.lineplot(data=df, x='Year', y='Life expectancy at birth (years)', hue='Country')
plt.show()
```

### Zimbabwe, despite being the Country with the lowest Life Expectancy, has had a drop (2 years decrease) on their population's life expectancy but since 2004, it had a major increase (15 years increase).
### Other countries had some increase. let's check that.

```python
others = df[df['Country'] != "Zimbabwe"]
LE_change = others.groupby('Country')['Life expectancy at birth (years)'].max() - others.groupby('Country')['Life expectancy at birth (years)'].min()
average_change = LE_change.mean()
print(LE_change)
print(average_change)
```

### On average, the other countries on average had a 3 years increase. Why would that be?
### Now, let's see how their GDP performed over those same years:

```python
sns.lineplot(data=df, x="Year", y="GDP", hue='Country')
plt.show()
```

### These countries have such a major difference in GDP. We should take a closer look at Zimbabwe:

```python
sns.lineplot(data=df, x="Year", y="GDP", hue='Country')
plt.ylim(2000000000, 300000000000)
```

### It seems like there is no correlation between GDP and Life Expectancy. What we have observed here is that a country with a low GDP is also a country with a low life expectancy. It also had a major GDP increase a few years _after_ it had an increase in life expectancy! Which could be a higher standard of living, health treatment or food, which could have an impact on their productivity aka. GDP! Further analysis should be done to check on that hypothesis.

### But from the graph, we can see how far in therms of GDP Zimbabwe is from the second lowest GDP country, Chile. That's a massive difference from the USA!
